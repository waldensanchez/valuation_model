### ajustar distribuciones

```python
# Create models from data
def best_fit_distribution(data, bins=200, ax=None):
    """Model data by finding best fit distribution to data"""
    # Get histogram of original data
    y, x = np.histogram(data, bins=bins, density=True)
    x = (x + np.roll(x, -1))[:-1] / 2.0

   
    DISTRIBUTIONS = [st.poisson, st.lognorm, st.norm, st.chi2, st.uniform, st.binom]

    # Best holders
    best_distribution = st.norm
    best_params = (0.0, 1.0)
    best_sse = np.inf

    # Estimate distribution parameters from data
    for distribution in DISTRIBUTIONS:

        # Try to fit the distribution
        try:
            # Ignore warnings from data that can't be fit
            with warnings.catch_warnings():
                warnings.filterwarnings('ignore')

                # fit dist to data
                params = distribution.fit(data)

                # Separate parts of parameters
                arg = params[:-2]
                loc = params[-2]
                scale = params[-1]

                # Calculate fitted PDF and error with fit in distribution
                pdf = distribution.pdf(x, loc=loc, scale=scale, *arg)
                sse = np.sum(np.power(y - pdf, 2.0))

                # if axis pass in add to plot
                try:
                    if ax:
                        pd.Series(pdf, x).plot(ax=ax)
                
                except Exception:
                    pass

                # identify if this distribution is better
                if best_sse > sse > 0:
                    best_distribution = distribution
                    best_params = params
                    best_sse = sse

        except Exception:
            pass

    return (best_distribution.name, best_params)
```

```python
# crear datframe con las distribuciones que mejor se ajustan a cada modelo
results = []

for i in ratios:
    distribution_params = best_fit_distribution(ratios_only_full[i])
    results.append({
        'Ratio': i,
        'DistributionParams': distribution_params
    })

df = pd.DataFrame(results)

#limpieza
df['DistributionType'] = df['DistributionParams'].apply(lambda x: x[0])
df['Parameters'] = df['DistributionParams'].apply(lambda x: x[1])
df['Parameters'] = df['Parameters'].apply(lambda x: tuple(round(param, 2) for param in x))

df.drop(columns=['DistributionParams'], inplace=True)
df.set_index('Ratio', inplace=True)

#crear un dataframe por cada distribucion
dataframes = {}
for distribution_type in df['DistributionType'].unique():
    dataframes[distribution_type] = df[df['DistributionType'] == distribution_type].copy()
```

```python
### Normal dist
dataframes["norm"][['Mean', 'Variance']] = dataframes["norm"]['Parameters'].apply(lambda x: pd.Series({'Mean': x[0], 'Variance': x[1]}))

# Drop the original 'Parameters' column
dataframes["norm"].drop(columns=['Parameters'], inplace=True)

dataframes["norm"]
```

```python
### lognormal dist
dataframes["lognorm"][['Mean', 'Variance']] = dataframes["lognorm"]['Parameters'].apply(lambda x: pd.Series({'Mean': x[0], 'Variance': x[1]}))

# Drop the original 'Parameters' column
dataframes["lognorm"].drop(columns=['Parameters'], inplace=True)

dataframes["lognorm"]
```

```python
### chi2 dist 
dataframes["chi2"][['k']] = dataframes["chi2"]['Parameters'].apply(lambda x: pd.Series({'k': x[0]}))

# Drop the original 'Parameters' column
dataframes["chi2"].drop(columns=['Parameters'], inplace=True)

dataframes["chi2"]
```
